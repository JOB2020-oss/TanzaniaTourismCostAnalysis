### Data Cleaning and Feature Engineering
Some of the data points such as <b>total male</b> and <b>total female</b> as well as <b>night_zanzibar</b> and <b>night_mainland</b> can be established to single featurs by doing <b>Feature Combination</b>
Also travel with seems to consisit of some null values initial filling the null value with mode ie frequent occuring value will be good idea also

```python
def process(data):
    ##FILLING travel_with with mode value
    data["travel_with"] = data["travel_with"].fillna(data["travel_with"].mode()[0])
    
    #FEATURE COMBINATION ON total_male AND total_female TO GET TOTAL NUMBER
    data["total_female"] = data["total_female"].fillna(method="ffill")
    data["total_male"] = data["total_male"].fillna(method="ffill")
    data["total"] = data["total_female"]+data["total_male"]
    
    #FEATURE COMBINATION ON night_mainland AND night_zanzibar TO GET TOTAL NUMBER
    data['nights'] = data['night_mainland']+data['night_zanzibar']
    data = data.drop(['total_female','total_male','night_mainland','night_zanzibar'],axis=1)
    data["total"] = data["total"].astype('int64')
    
    #Dropping Tour_ID also
    iD = data["Tour_ID"]
    data = data.drop("Tour_ID",axis=1)
    
    return data
train,test = process(train),process(test)
```
